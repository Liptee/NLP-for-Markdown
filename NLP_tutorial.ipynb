## Stemming

```python
from nltk.stem import PorterStemmer
pst = PorterStemmer()
```

```python
words_to_stem=["give","giving", "given", "gave", "gaves", "gaved"]
for words in words_to_stem:
    print(f"{words}: {pst.stem(words)}")
```

```python
from nltk.stem import LancasterStemmer
lst = LancasterStemmer()
for word in words_to_stem:
    print(f"{word}: {lst.stem(word)}")
```

```python
from nltk.stem import SnowballStemmer
sbst_ru = SnowballStemmer('russian')
```

```python
words_to_stem_ru = ['обычай', 'обычный', "обычная", "обучаемый"]
for word in words_to_stem_ru:
    print(f"{word}: {sbst_ru.stem(word)}")
```

## Named Entity Recognition

```python
from nltk import ne_chunk
```

```python
NE_sent = "Development of HTTP was initiated by Tim Berners-Lee at CERN in 1989 and summarized in a simple document describing the behavior of a client and a server using the first HTTP protocol version that was named 0.9."
```

```python
NE_tokens = word_tokenize(NE_sent)
NE_tags = nltk.pos_tag(NE_tokens)
```

```python
NE_NER = ne_chunk(NE_tags)
print(NE_NER)
```

## Syntax Tree/Chunking

```python
new = "The big cat ate the little mouse who was after fresh cheese"
new_tokens = nltk.pos_tag(word_tokenize(new))
new_tokens
```

```python
grammar_np = r"NP: {<DT>?<JJ>*<NN>}"
```

```python
chunk_parser = nltk.RegexpParser(grammar_np)
```

```python
chunk_result = chunk_parser.parse(new_tokens)
chunk_result
```
